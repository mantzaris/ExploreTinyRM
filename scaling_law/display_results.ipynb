```python
import os
import pandas as pd
import glob

# Find all metrics.csv files in results subfolders
results_dir = os.path.join(os.getcwd(), 'results')
metrics_files = glob.glob(os.path.join(results_dir, '**', 'metrics.csv'), recursive=True)

# Read and concatenate all results
all_results = []
for metrics_file in metrics_files:
    df = pd.read_csv(metrics_file)
    # Add config info if available
    config_path = os.path.join(os.path.dirname(metrics_file), 'config.json')
    if os.path.exists(config_path):
        import json
        with open(config_path, 'r') as f:
            config = json.load(f)
        for k, v in config.items():
            df[k] = v
    all_results.append(df)

if all_results:
    results_df = pd.concat(all_results, ignore_index=True)
    display(results_df)
else:
    print("No results found in results/ directory.")
```
